```python
batchName = 'v0_batch0'

with open(batchName + '/' + batchName + '_batch.json', 'r') as f:
    batchInfo = json.load(f) 
            
params = batchInfo['batch']['params']   
print(params)  
gex = []
I = []
for name in params:
    for value in name['values']:
        if np.size(name['label']) < 2:
            print(name['label'],value)
            gex.append(value)
            
        else:
            print('I',value)
            I.append(value)
```

```python
data = {}
paramsNameList = []

for ii,p1 in enumerate(gex):
    for jj,p2 in enumerate(I):
        paramsName  = 'gex' + str(p1) + '_' + 'I' + str(p2)
        paramsNameList.append(paramsName)
        
        data[paramsName] = {}
        for label in ['CV', 'rate', 'sync']:
            with open(batchName + '/' + batchName + '_' + str(ii) + '_' + str(jj) +'_' + label +'.json', 'r') as f:
                stats = json.load(f) 
                data[paramsName][label] = stats['statData']
```
